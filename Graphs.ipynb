### This notebook is used to plot the results from the different experiments

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#df_mrr = pd.read_csv('df_mrr_chroma.csv')

df_mrr = pd.read_csv('df_mrr_fft.csv')

#df_mrr = pd.read_csv('df_mrr_fft_amplitude_bin_reduction.csv')

#df_mrr = pd.read_csv('df_mrr_fft_my_data.csv')

    
dfsp=df_mrr.ftype=='spectrum'
df_spec=df_mrr[dfsp]

dfsp=df_mrr.ftype=='degree'
df_degree=df_mrr[dfsp]

dfsp=df_mrr.ftype=='degree distribution'
df_degreedist=df_mrr[dfsp]


plt.figure(figsize=[6,4])
plt.plot(df_spec[df_spec.dtype=='euclidean'].SNR,df_spec[df_spec.dtype=='euclidean'].MRR, 'x:b')
plt.plot(df_degree[df_degree.dtype=='euclidean'].SNR,df_degree[df_degree.dtype=='euclidean'].MRR, 'o:r')
plt.plot(df_degreedist[df_degreedist.dtype=='euclidean'].SNR,df_degreedist[df_degreedist.dtype=='euclidean'].MRR, 'd:g')

plt.xlabel('SNR')
plt.ylabel('MRR')
plt.title('euclidean')

# for chroma y lim use this
#plt.ylim([0,.2])

# for fft y lim use this
plt.ylim([0,1])

plt.grid('both')
#change for different experiments
#plt.legend(['chroma','degree','degreedst'])
plt.legend(['spectrum','degree','degreedst'])
#plt.savefig('Graphs/euclidean_fft_with_amplitude_bin_reduction.png')
plt.show()

plt.figure(figsize=[6,4])
plt.plot(df_spec[df_spec.dtype=='cosine'].SNR,df_spec[df_spec.dtype=='cosine'].MRR, 'x:b')
plt.plot(df_degree[df_degree.dtype=='cosine'].SNR,df_degree[df_degree.dtype=='cosine'].MRR, 'o:r')
plt.plot(df_degreedist[df_degreedist.dtype=='cosine'].SNR,df_degreedist[df_degreedist.dtype=='cosine'].MRR, 'd:g')
#plt.plot(df_chroma[df_chroma.dtype=='cosine'].SNR,df_chroma[df_chroma.dtype=='cosine'].MRR, '*:k')


plt.xlabel('SNR')
plt.ylabel('MRR')
plt.title('Cosine')

# for chroma y lim use this
#plt.ylim([0,.4])

# for fft y lim use this
plt.ylim([0,1])

plt.grid('both')
#change for different experiments
#plt.legend(['chroma','degree','degreedst'])
plt.legend(['spectrum','degree','degreedst'])
#plt.savefig('Graphs/cosine_fft_with_amplitude_bin_reduction.png')

plt.show()

```
